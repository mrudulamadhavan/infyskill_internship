## **Create a 4*4 subplot and you will be having 16 subplots on the canvas**

- Take some polynomial functions and trigonometric functions and plot them on subplots.


```python
import numpy as np
import matplotlib.pyplot as plt

# Function definitions
def quadratic(x):
    return x**2

def cubic(x):
    return x**3

def sine(x):
    return np.sin(x)

def cosine(x):
    return np.cos(x)

# Create a 4x4 subplot
fig, axes = plt.subplots(nrows=4, ncols=4, figsize=(12, 12))

# Flatten the 2D array of subplots for easier indexing
axes = axes.flatten()

# Define x values
x_values = np.linspace(-2*np.pi, 2*np.pi, 100)

# Plot polynomial and trigonometric functions on subplots with function names
functions = [quadratic, cubic, sine, cosine]

for i, ax in enumerate(axes):
    y_values = functions[i % len(functions)](x_values)
    function_name = functions[i % len(functions)].__name__
    ax.plot(x_values, y_values)
    ax.set_title(f'{function_name}')

# Adjust layout to prevent overlapping
plt.tight_layout()
plt.show()
```
