#### **Take flights dataset and penguins dataset and create some visualizations using countplot and relplot with different parameters.**

```python
import seaborn as sns
import matplotlib.pyplot as plt
import seaborn as sns

# Load the "penguins" dataset
penguins_data = sns.load_dataset("penguins")
penguins_data
```

```python
# Create a countplot for the "penguins" dataset
plt.figure(figsize=(12, 6))
sns.countplot(x="species", data=penguins_data)
plt.title("Count of Penguins by Species")
plt.show()

```

```python
#  Pairplot for Penguins Data
sns.pairplot(penguins_data, hue="species", height=3, aspect=1.2)
plt.suptitle("Pairplot for Penguin Dataset", y=1.02)
plt.show()
```

```python
# Create a relplot for the "penguins" dataset
sns.relplot(x="flipper_length_mm", y="bill_length_mm", data=penguins_data, kind="scatter", hue="species", size="body_mass_g")
plt.title("Penguin Flipper Length vs. Bill Length")
plt.show()
```

```python
# Boxplot for Penguins Data
plt.figure(figsize=(10, 6))
sns.boxplot(x="species", y="body_mass_g", data=penguins_data, hue="island")
plt.title("Body Mass Distribution of Penguins by Species and Island")
plt.show()
```

```python
# Load the "flights" dataset
flights_data = sns.load_dataset("flights")
flights_data
```

```python
# Lineplot for Flights Data
plt.figure(figsize=(12, 6))
sns.lineplot(x="month", y="passengers", hue="year", data=flights_data, marker='o')
plt.title("Passenger Count Over Months for Flights Dataset")
plt.show()
```

```python
# Create a countplot for the "flights" dataset
plt.figure(figsize=(12, 4))
sns.countplot(x="month", data=flights_data)
plt.title("Count of Flights per Month")
plt.show()
```

```python
#  Heatmap for Flights Data
plt.figure(figsize=(12, 8))
flights_pivot = flights_data.pivot_table(index='month', columns='year', values='passengers')
sns.heatmap(flights_pivot, cmap='YlGnBu', annot=True, fmt='d', linewidths=.5)
plt.title("Passenger Count Heatmap for Flights Dataset")
plt.show()
```

```python
# Create a relplot for the "flights" dataset
sns.relplot(y="passengers", x="month", data=flights_data, kind="scatter", hue="year", col="year")
plt.suptitle("Passenger Counts Over Months (Split by Year)", y=1.8)
plt.show()
```
