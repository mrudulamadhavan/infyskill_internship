```python
# importing libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Q1) Plot a graph for cosine function in the interval [0, 10 * pi ]

x = np.linspace(0, 10 * np.pi, 1000)
y = np.cos(x)

plt.plot(x, y)
plt.title('Cosine Function')
plt.xlabel('x')
plt.ylabel('cos(x)')
plt.grid(True)
plt.show()
```

```python
# Q2) Plot a graph for logarithmic function and how does it look like ?

x = np.linspace(0.1, 5, 100)
y = np.log(x)

plt.plot(x, y)
plt.title('Logarithmic Function')
plt.xlabel('x')
plt.ylabel('log(x)')
plt.grid(True)
plt.show()
```

### **Q) Can you change the numpy array size after declaration? Explain briefly your views.**
No, the size of a NumPy array is fixed upon creation. Unlike Python lists, NumPy arrays have a fixed size. If you want to change the size, you need to create a new array. This is because NumPy arrays are designed for efficient numerical operations, and a fixed size allows for optimizations that wouldn't be possible if the size were variable.

### **Q) Write a function for finding the nth prime number .**

```python
def is_prime(num):
    if num < 2:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

def nth_prime(n):
    primes = []
    i = 2
    while len(primes) < n:
        if is_prime(i):
            primes.append(i)
        i += 1
    return primes[-1]

# Example usage:
print(nth_prime(5))  # Output: 11
```

### **Q) Discuss about the differences between numpy array and python list.**

* **Memory Efficiency:** NumPy arrays are more memory-efficient than Python lists. NumPy uses a fixed-type array, which allows for more efficient storage and operations.

* **Performance:** NumPy operations are faster than equivalent operations on Python lists, especially for large datasets, due to optimized C and Fortran libraries.

* **Convenience:** Lists are more flexible as they can hold elements of different data types. NumPy arrays, on the other hand, are homogeneous and require elements to be of the same type.

* **Functionality:** NumPy provides a wide range of mathematical functions and operations that are not available in the standard Python list.

* **Broadcasting:** NumPy allows for broadcasting, a powerful feature that simplifies operations on arrays of different shapes.

* **Ease of Use:** For simple, general-purpose tasks, lists might be more convenient. NumPy is particularly beneficial for numerical and scientific computing tasks.

Choosing between a list and a NumPy array depends on the specific requirements of your task. If you're working with numerical data and need performance, NumPy is often the better choice. For general-purpose data manipulation and storage of mixed types, lists might be more appropriate.
